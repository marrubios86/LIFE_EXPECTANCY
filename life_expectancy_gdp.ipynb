```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_countries = pd.read_csv('all_data.csv')
data_countries.head(20)
```

```python
data_countries.info()
```

```python
print(data_countries.Country.unique())
```

```python
data_countries.rename(columns={'Life expectancy at birth (years)': 'Lifexp'}, inplace=True)
Life_exp_years = data_countries.Lifexp
gdp = data_countries.GDP
```

```python
sns.distplot(data_countries.GDP)
plt.xlabel('GDP')
plt.show()
```

```python
sns.distplot(data_countries.Lifexp)
plt.xlabel('Life expectancy in years')
plt.show()
```

```python
sns.boxplot(x='Country', y=Life_exp_years, data=data_countries, palette='pastel')
plt.xticks(rotation=45)
plt.show()
```

```python
gdp_mean = data_countries.groupby('Country').GDP.mean().reset_index()
gdp_mean
```

```python
plt.bar(data_countries.Country, data_countries.GDP)
plt.xlabel('Country')
plt.ylabel('GDP')
plt.title('GDP in every country')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()
plt.show()
```

```python
lifexp_mean = data_countries.groupby('Country').Lifexp.mean().reset_index()
lifexp_mean
```

```python
plt.bar(data_countries.Country, data_countries.Lifexp)
plt.xlabel('Country')
plt.ylabel('Life expectancy in years')
plt.title('Life expectancy in every country')
plt.xticks(rotation=45, ha='right') 
plt.ylim(40)
plt.tight_layout()
plt.show()
```

```python
sns.lineplot(x=data_countries.Year, y=data_countries.GDP, hue=data_countries.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("GDP in Trillions of U.S. Dollars")
plt.show()
```

```python
sns.lineplot(x=data_countries.Year, y=data_countries.Lifexp, hue=data_countries.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("Life expectancy in years")
plt.show()
```

```python
sns.scatterplot(x = 'Lifexp', y = 'GDP', hue='Country', data = data_countries)      
plt.show()
```

Conclusions:
As we can see, there is no clear relationship between life expectancy and GDP, because countries like Chile, that have a low GDP, have a higher life expectancy than countries with a higher GDP like China. In the case of Zimbabwe, the super low GDP could explain their low life expectancy, but it could also be dued to other facts, which be aren't analizing here.
